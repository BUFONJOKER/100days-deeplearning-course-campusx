```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
or_data = pd.DataFrame({
    'input_1':[1,1,0,0],
    'input_2':[1,0,1,0],
    'output':[1,1,1,0]
})

and_data = pd.DataFrame({
    'input_1':[1,1,0,0],
    'input_2':[1,0,1,0],
    'output':[1,0,0,0]
})

xor_data = pd.DataFrame({
    'input_1':[1,1,0,0],
    'input_2':[1,0,1,0],
    'output':[0,1,1,0]
})
```

```python
sns.scatterplot(data=or_data,x='input_1', y='input_2', hue='output')
plt.title("OR Data\n")
plt.show()
```

```python
sns.scatterplot(data=and_data,x='input_1', y='input_2', hue='output')
plt.title("AND Data\n")
plt.show()
```

```python
sns.scatterplot(data=xor_data,x='input_1', y='input_2', hue='output')
plt.title("XOR Data\n")
plt.show()
```

```python
from sklearn.linear_model import Perceptron

model_1 = Perceptron()
model_2 = Perceptron()
model_3 = Perceptron()
```

```python
model_1.fit(and_data.iloc[:,0:2].values, and_data.iloc[:,-1].values)
model_2.fit(or_data.iloc[:,0:2].values, or_data.iloc[:,-1].values)
model_3.fit(xor_data.iloc[:,0:2].values, xor_data.iloc[:,-1].values)
```

```python
x=np.linspace(-1,1,5)
y=-x+1
model_1.coef_,model_1.intercept_
```
